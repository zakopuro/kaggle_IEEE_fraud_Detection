```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sub = pd.read_csv('/Users/zakopuro/Code/python_code/kaggle/IEEE_Fraud_Detection/output/055_sub_lgb.csv')
```

```python
pre_sub = pd.read_csv('../output/085_sub_lgb.csv')
```

```python
def plot_isFraud(df):
    plt.figure(figsize=(10,30))
    for i in range(5):
        plt.subplot(5,1,i+1)
        x = df[int(len(df)*0.2)*i:int(len(df)*0.2)*(i+1)]['isFraud']
        plt.hist(x,bins=30)
        plt.ylim(0,10000)
        plt.title(str((i)*20)+'-'+str((i+1)*20))
```

```python
plot_isFraud(pre_sub)
```

```python
plot_isFraud(sub)
```
